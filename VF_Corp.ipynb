```python
%cd /content/drive/MyDrive/

# Import libraries.
import pandas as pd
import numpy as np
import re

df = pd.read_excel('Test raw file.xlsx')

# Print the DataFrame.
df
```

```python
# check data types
# no null info
df.info()
df['Factory Name'] = df['Factory Name'].astype(str)
```

```python
# Question 1.1,drop and recheck for duplicated values
df = df.drop_duplicates()
df.duplicated()
```

```python
#Question 1.2, clean up factory name
def remove_symbols(factory_name):
  clean_factory_name = re.sub(r'[^a-zA-Z0-9]', '', factory_name)

  return clean_factory_name

df['Factory Name'] = df['Factory Name'].apply(remove_symbols)
df
```

```python
# Question 1.3 Split brand using delimiter and drop old column
df[['Brand1','Brand2','Brand3']] = df['Brand'].str.split("|", expand=True)
df.drop('Brand',axis=1,inplace=True)
df
```

```python
# Question 1.4 Standardize the In Scope data
df['In Scope'] = df['In Scope'].replace({'Yes': 'Y', 'No': 'N'})
df
```

```python
#Question 1.5, remove In Scope = 'N'
df = df[df['In Scope'] != 'N']
df
```

```python
#Question 1.6 Save to excel
df.to_excel('Gabriel Cleaned Test.xlsx', index=False)
```
